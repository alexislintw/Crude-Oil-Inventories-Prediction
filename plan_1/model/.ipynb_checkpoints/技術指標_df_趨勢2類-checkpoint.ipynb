```python
import numpy as np
import talib
import pandas as pd
#import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../dataset/^TWII.csv")
df = df.dropna()
close = df['Adj Close'].values
df = df.drop(['Date','Open','High','Low','Close','Volume'], 1)
#close = [float(x) for x in df['Adj Close']]
close = np.array(close)
#df.head(10)
```

```python
trend = []
for idx, c in enumerate(close):
    if idx < len(close) - 1:
        re = close[idx+1] - c
        tr = 0
        if re > 0:
            tr = 1
    else :
        tr = 0
    trend.append(tr)

df['Trend'] = np.array(trend)
df.head(10)
```

```python
n = 20
df['SMA'] = talib.SMA(close,timeperiod=n)
df['ROC'] = talib.ROC(close,timeperiod=n)
df['RSI'] = talib.RSI(close,timeperiod=n)
df['MACD'],df['MACDsignal'],df['MACDhist'] = talib.MACD(np.array(close),fastperiod=6, slowperiod=12, signalperiod=9)
df = df.dropna()
df = df.drop(['MACDsignal','MACDhist'], 1)
df.head(10)
```
